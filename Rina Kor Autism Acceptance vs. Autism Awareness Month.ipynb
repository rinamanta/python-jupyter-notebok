##    Introduction 

In January of this year, a company called Austimming Tools and Toys hired Gib Atad for data analysis of their market. Austimming sells products that can be used by autistic people for sensory stimulation ("stimming"). They also sell products that help autistic people with organization and time management. 

In the past--like most companies in the autism market--Austimming's advertising focused exclusively on the parents of autistic children--not on autistic people themselves. However, Austimming has heard from hundreds of autistic people who buy the products for their own use. So, Austimming now wants to diversify their advertising. They want to write campaigns that appeal to both autistic people and to the parents of autistic people (with recognition that those two groups have a large area of intersection). 


In the past, Austimming has offered special promotions during the month of April, and has used the term "Autism Awareness Month" when marketing those promotions. Autistic adults have told the company that the term "Autism Awareness Month" offends them. That's because Autism Awareness Month was started by parents' groups that wanted "awareness" in order to find a cure. The vast majority of autistic adults do not see themselves as diseased; they see themselves as different, not less. Many autistic customers told Austimming that they prefer the term "Autism Acceptance Month" instead.

To provide guidance and insight to Austimming, I gathered tweets made during the month of April, 2021. I did comparative analysis on 1000 tweets containing the terms "autism" and "acceptance", and 1000 tweets containing the terms "autism" and "awareness". I present some of my findings below.

## Conclusion 

In order for Austimming to diversify their marketing--and subsequently increase the size of their market--they need to use language that appeals to the broader autistic community. 

"Acceptance" is gaining traction even in the community of parents of autistic children, a group that was the catalyst of the original Autism Awareness month. "Acceptance" is definitely the term preferred by autistic adults themselves. Autistic adults also emphasize community. They want to focus on the positive aspects of being autistic. I advise Austimming to showcase their products in a positive light, to highlight the happiness they bring to autistic people of any age.
